```python
import pandas as pd
```

```python
valores_ausentes = ['NaN', "***"]
df = pd.read_csv("ocorrencia.csv", sep=";", parse_dates=["ocorrencia_dia", "divulgacao_dia_publicacao"], dayfirst=True, na_values=valores_ausentes)
df.head(10)
```

```python
df.dtypes
```

```python
display(df.loc[1:3])
```

```python
#Ocorrencias com mais de 10 recomendações
filtro = df.total_recomendacoes > 10
df.loc[filtro]
```

```python
#Ocorrencias com mais de 10 recomendações
filtro = df.ocorrencia_classificacao == "INCIDENTE GRAVE"
df.loc[filtro, ["ocorrencia_cidade", "total_recomendacoes"]]
```

```python
#Ocorrencias com classificação = acidente grave
filtro = df.total_recomendacoes > 10
df.loc[filtro, ["ocorrencia_cidade", "total_recomendacoes"]]
```

```python
#Ocorrencias com classificação = acidente grave no estado de SP
filtro1 = df.total_recomendacoes > 10
filtro2 = df.ocorrencia_uf == "SP"
df.loc[filtro1 & filtro2, ["ocorrencia_cidade", "total_recomendacoes"]]
```

```python
#Ocorrencias com classificação = incidente grave ou incidente no estado de SP
filtro1 = (df.ocorrencia_classificacao == "INCIDENTE GRAVE") | (df.ocorrencia_classificacao == "INCIDENTE")
filtro2 = df.ocorrencia_uf == "SP"
df.loc[filtro1 & filtro2, ["ocorrencia_cidade", "total_recomendacoes"]]
```

```python
#Ocorrencias com classificação = incidente grave ou incidente no estado de SP usando o isin()
filtro1 = df.ocorrencia_classificacao.isin(["INCIDENTE GRAVE", "INCIDENTE"])
filtro2 = df.ocorrencia_uf == "SP"
df.loc[filtro1 & filtro2, ["ocorrencia_cidade", "total_recomendacoes"]]
```

```python
#ocorencia cuja cidades comecem coma letra c
filtro = df.ocorrencia_cidade.str[0] == "C"
df.loc[filtro]
```

```python
#ocorencia cuja cidades terminam com a letra a
filtro = df.ocorrencia_cidade.str[-1] == "A"
df.loc[filtro]
```

```python
#ocorencia cuja cidades terminam com ma
filtro = df.ocorrencia_cidade.str[-2:] == "MA"
df.loc[filtro]
```

```python
#ocorrencias do ano de 2015
filtro = df.ocorrencia_dia.dt.year == 2015
df.loc[filtro]
```

```python
#ocorrencias do ano de 2015  e mes 12
filtro1 = df.ocorrencia_dia.dt.year == 2015
filtro2 = df.ocorrencia_dia.dt.month == 12
df.loc[filtro1 & filtro2]
```

```python
#ocorrencias do ano de 2015  e mes 12 e dias entre 3 e 8
filtro_ano = df.ocorrencia_dia.dt.year == 2015
filtro_mes = df.ocorrencia_dia.dt.month == 12
filtro_dias = (df.ocorrencia_dia.dt.day >= 3) & (df.ocorrencia_dia.dt.day <= 8)
df.loc[filtro_ano & filtro_mes & filtro_dias]
```

```python
df.ocorrencia_dia.astype(str) + " - " + df.ocorrencia_hora #concatena como string
```

```python
df["ocorrencia_dia_hora"] = pd.to_datetime(df.ocorrencia_dia.astype(str) + " - " + df.ocorrencia_hora) #concatena como data e hora
df.ocorrencia_dia_hora
```

```python
#novo dataframe com filtro de ocorrencias do ano de 2015  e mes 12 e dias entre 3 e 8
filtro_ano = df.ocorrencia_dia.dt.year == 2015
filtro_mes = df.ocorrencia_dia.dt.month == 12
filtro_dias = (df.ocorrencia_dia.dt.day >= 3) & (df.ocorrencia_dia.dt.day <= 8)
dfnovo = df.loc[filtro_ano & filtro_mes & filtro_dias]
dfnovo
```
